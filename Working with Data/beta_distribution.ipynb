## <span style="font-size: 18pt; color: goldenrod; font-weight:bold">Beta Distribution</span>
### Overview

The <span style="font-size: 11pt; color: goldenrod; font-weight:bold">Beta Distribution</span> is a <u>continuous probability distribution</u> defined on the interval $[0, 1]$. It is characterized by two positive shape parameters, denoted as $\alpha$ and $\beta$, which determine the shape of the distribution. The distribution is used to model random variables that are constrained to lie within the $[0, 1]$ range, making it **particularly useful for modeling proportions, probabilities, and other quantities that are bounded**.
***
The beta distribution was first introduced by the mathematician and statistician **Thomas Bayes**. However, its properties and applications were further explored and developed by other statisticians, such as **Simon Newcomb** and **A.M. Walker**, in the late 19th and early 20th centuries.
***
The Beta distribution plays a <u>crucial role in Bayesian analysis</u> and various applications where modeling probabilities, proportions, and bounded random variables is essential. Its flexibility and conjugate properties make it an indispensable tool for statisticians, data analysts, and researchers across diverse fields.
### Important Properties:

- <span style="font-size: 11pt; color: steelblue; font-weight:bold">Probability Density Function (PDF)</span>: 

$$ f(x; \alpha, \beta) = \frac{1}{B(\alpha, \beta)} \cdot x^{\alpha-1} \cdot (1-x)^{\beta-1} $$

where:
    - $(x)$ is the value of the random variable in the interval $[0, 1]$, $\alpha$ and $\beta$ are the positive shape parameters, $B(\alpha, \beta)$ is the beta function, a normalization constant.

- <span style="font-size: 11pt; color: steelblue; font-weight:bold">Mean</span>: $$\frac{\alpha}{\alpha + \beta}$$
- <span style="font-size: 11pt; color: steelblue; font-weight:bold">Variance</span>: $$\frac{\alpha \cdot \beta}{(\alpha + \beta)^2 \cdot (\alpha + \beta + 1)}$$

### Alpha and Beta parameters:
Alpha $\alpha$ and beta $\beta$ are the two shape parameters that define the distribution. These parameters play a crucial role in shaping the characteristics of the beta distribution, determining its spread, skewness, and the location of the mode. They are what gives the Beta De 


- **Alpha**: Alpha represents the first shape parameter of the beta distribution. It is a positive real number that <u>controls the weight and concentration of probability density around the lower end of the distribution (near 0)</u>. Larger values of α result in more probability mass being concentrated towards 1, while smaller values of α lead to more probability mass concentrated towards 0.

- **Beta**: Beta represents the second shape parameter of the beta distribution. Similar to α, β is also a positive real number. <u>It controls the weight and concentration of probability density around the upper end of the distribution (near 1)</u>. Larger values of β result in more probability mass being concentrated towards 0, while smaller values of β lead to more probability mass concentrated towards 1.

### Applications:

The beta distribution has numerous applications in various fields, including:
- **Bayesian statistics**: It serves as a **conjugate prior** for the binomial and multinomial likelihood functions, simplifying the computations of posterior distributions.
- **Modeling proportions and probabilities**: Used to represent uncertainties in success rates, conversion rates, and proportions of certain events, like surveys or experiments
- **Reliability engineering**: Used to model the reliability of components and systems.
- **Risk analysis**: Modeling the probability of default in credit risk analysis.
- **A/B testing**: Employed to analyze and compare the performance of different variants or designs.


```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import beta

# Define the shape parameters
alpha = 2
beta_ = 5

# Create a range of values from 0 to 1
x = np.linspace(0, 1, 100)

# Calculate the corresponding PDF values
pdf = beta.pdf(x, alpha, beta_)

# Plot the beta distribution
plt.plot(x, pdf)

plt.xlabel('x')
plt.xticks(np.arange(0,1.1,0.1))
plt.ylabel('Probability Density Function')
plt.grid(True, ls='dashed', alpha=0.3)
plt.title(f'Beta Distribution (α={alpha}, β={beta_})')

plt.show()
```
