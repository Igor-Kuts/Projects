# Data Statistics

A **statistic** is a numerical value or measurement that describes a characteristic or property of a dataset. It is derived from the analysis or summary of the data and is used to gain insights or make inferences about a population based on the sample data.

**Statistics** are used to summarize and describe different aspects of the data, such as central tendency (mean, median, mode), dispersion (variance, standard deviation, range), correlation, regression, probability, and many other measures and methods.

The main descriptive statistics of the data are measures that provide a summary of the data's distribution and central tendencies.

They are:

- Mean
- Median
- Mode
- Range
- Variance
- Standard deviation
- Interquartile Range
***
Below is an overview of each one of them with brief overview, formulas, and visualizations using plots for better understanding.
***

```python
# Importing libraries
import numpy as np
import matplotlib.pyplot as plt
```

## Mean:

The mean is the average value of a dataset, calculated by summing all the values and dividing by the number of observations.

The mean is sensitive to extreme values and can be affected by outliers.

Formula:
$$\bar{x} = \frac{1}{n}\sum_{i=1}^{n} x_i$$

```python
data = [0, 2, 3, 2, 4, 4, 3, 4, 3, 9, 7, 7]

# Computing mean using numpy library
mean = np.mean(data)
print('\nMean of the data:', mean)

# Plotting the data and the mean as a line
plt.figure(figsize=(8, 3))
plt.plot(data, label='Data', marker='o')

# Add mean line to a plot
plt.vlines(mean, 0, max(data), color='orange', linewidth=1, label='Mean')

# Customize the plot
plt.xlabel('Index')
plt.ylabel('Value')
plt.title('Mean')
plt.grid(True, ls='dashed', alpha=0.5)
plt.legend()

# Display the plot
plt.legend()

plt.show()
```

## Median:

The median is the middle value in a dataset when it is sorted. If the dataset has an even number of observations, the median is the average of the two middle values.

The median is a robust measure of central tendency that is less influenced by extreme values compared to the mean.

Formula: 
$$ \text{Median} = \begin{cases} \text{Middle value of the sorted dataset} & \text{if } n \text{ is odd} \\ \frac{{\text{Sum of the two middle values of the sorted dataset}}}{2} & \text{if } n \text{ is even} \end{cases}
$$

```python
data = np.array([0, 2, 3, 2, 4, 4, 3, 4, 3, 9, 7, 7])

# Computing median using NumPy
median = np.median(data)
print('\nMedian of the data:', median)

# Plotting the data and the median as a line
plt.figure(figsize=(8, 3))
plt.plot(data, label='Data', marker='o')

# Add the median line to the plot
plt.vlines(median, 0, max(data), color='orange', linewidth=1, label='Median')

# Customize the plot
plt.grid(True, ls='dashed', alpha=0.5)
plt.xlabel('Index')
plt.ylabel('Value')
plt.title('Median')
plt.legend()


# Display the plot
plt.show()
```

## Mode:

The mode represents the most frequently occurring value(s) in a dataset.  

A dataset can have multiple modes or no mode at all (when all values are unique).


```python
data = [0, 2, 3, 2, 4, 4, 3, 4, 3, 9, 7, 7]

# Computing the mode using NumPy
unique_values, counts = np.unique(data, return_counts=True)
mode_indices = np.argmax(counts)
mode = unique_values[mode_indices]
print('\nMode of the data:', mode)

# Plotting the data and the Mode as a line
plt.figure(figsize=(8, 3))
plt.plot(data, label='Data', marker='o')

# Add mode line to the plot
plt.vlines(mode, 0, max(data), color='orange', linewidth=1, label='Mode')


# Customizing the plot
plt.grid(True, ls='dashed', alpha=0.5)
plt.title('Mode')
plt.xlabel('Index')
plt.ylabel('Value')
plt.legend()

# Display the plot
plt.show()
```

## Range:

The range is the difference between the maximum and minimum values in a dataset.

The range is affected by outliers and provides a rough estimate of the spread of the data.

Formula: $$\text{Range} = \text{max}(x_i) - \text{min}(x_i)$$

```python
data = [0, 2, 3, 2, 4, 4, 3, 4, 3, 9, 7, 7]

# Calculate the range using NumPy
data_range = max(data) - min(data)
print('\nRange of the data:', data_range)

# Prepare the plot
fig, ax = plt.subplots(figsize=(8, 3))

# Plot the data points
ax.plot(data, marker='o', linestyle='-', label='Data')

# Add error bars to represent the range
ax.errorbar(x=range(len(data)), y=data, yerr=data_range/2, \
            fmt='none', c='orange', capsize=3, label='Range')

# Customize the plot
plt.grid(True, ls='dashed', alpha=0.5)
ax.set_title('Range')
plt.xlabel('Index')
plt.ylabel('Value')
plt.legend()

# Display the plot
plt.show()
```

## Variance:

Variance measures the average squared deviation from the mean and provides an indication of the spread or dispersion of the data.

Variance is always a non-negative value.

Formula: $$\text{Var}(x) = \frac{1}{n} \sum_{i=1}^{n} (x_i - \bar{x})^2$$

```python
data = [0, 2, 3, 2, 4, 4, 3, 4, 3, 9, 7, 7]

# Calculate the variance using NumPy
variance = np.var(data)
print('Variance of the data:', variance)

# Prepare the plot
fig, ax = plt.subplots(figsize=(8, 3))

# Plot the data points
ax.plot(data, marker='o', linestyle='-', label='Data')

# Add error bars to represent the variance
ax.errorbar(x=range(len(data)), y=data, yerr=np.sqrt(variance), \
            fmt='none', c='orange', capsize=3, label='Variance')

# Customize the plot
plt.grid(True, ls='dashed', alpha=0.5)
ax.set_title('Variance')
plt.xlabel('Index')
plt.ylabel('Value')
plt.legend()

# Display the plot
plt.show()
```

## Standard Deviation:

The standard deviation is the square root of the variance and provides a measure of the average distance between each data point and the mean.

Standard deviation is often used in conjunction with the mean to describe the distribution of data.

Formula: $$\sigma = \sqrt{\text{Var}(x)}$$

```python
data = [0, 2, 3, 2, 4, 4, 3, 4, 3, 9, 7, 7]

# Calculate the standard deviation with NumPy
data_std = np.std(data)
print('Standard deviation of the data:', data_std)

# Prepare the plot
fig, ax = plt.subplots(figsize=(8, 3))

# Plot the data points
ax.plot(data, marker='o', linestyle='-', label='Data')

# Add error bars to represent the standard deviation
ax.errorbar(x=range(len(data)), y=data, yerr=data_std, \
            fmt='none', c='orange', capsize=3, label='Standard Deviation')

# Customize the plot
plt.grid(True, ls='dashed', alpha=0.5)
ax.set_title('Standard Deviation')
plt.xlabel('Index')
plt.ylabel('Value')
plt.legend()

# Display the plot
plt.show()
```

## Interquartile Range (IQR):
The IQR is the range between the first quartile (25th percentile) and the third quartile (75th percentile) and represents the spread of the central 50% of the data.

IQR is used to identify outliers and describe the variability of the middle half of the dataset.

Formula: $$\text{IQR} = Q_3 - Q_1$$

IQR can easily be visualized with the botplot. The length of the box in a box plot corresponds to the IQR, with the box spanning from the first quartile (Q1) to the third quartile (Q3).

```python
data = [0, 2, 3, 2, 4, 4, 3, 4, 3, 9, 7, 7]

# Calculate the quartiles using NumPy
q1 = np.percentile(data, 25)
q3 = np.percentile(data, 75)

# Calculate the IQR
iqr = q3 - q1
print('\nIQR of the data:', iqr)

# Prepare the plot
fig, ax = plt.subplots(figsize=(3, 8))

# Create a vertical box plot of the data
ax.boxplot(data, vert=True, widths=0.7, patch_artist=True, medianprops={'color': 'red'}, boxprops={'color': 'blue', 'facecolor': 'lightblue'}, whiskerprops={'color': 'blue'}, capprops={'color': 'blue'})

# Add text annotation for the IQR
ax.text(0.95, 0.9, f'IQR = {iqr:.2f}', transform=ax.transAxes, ha='right', va='top')

# Customize the plot
plt.grid(True, ls='dashed', alpha=0.5)
ax.set_title('Interquartile Range')
ax.set_xlabel('Data')
ax.set_ylabel('Value')

# Display the plot
plt.show()
```
