## <span style="font-size: 18pt; color: goldenrod; font-weight:bold">Chi-Squared Distribution</span>

### Overview:
The <span style="font-size: 11pt; color: goldenrod; font-weight:bold">Chi-Squared Distribution</span> is a continuous probability distribution that arises in statistics and probability theory. It is a special case of the gamma distribution and is widely <u>used in hypothesis testing</u> and in <u>constructing confidence intervals for population variances when dealing with normally distributed data</u>.

 The **degrees of freedom parameter**, $k$, determines the shape of the distribution. Higher values of $k$ result in distributions that are more bell-shaped and concentrated around their mean.

***
The chi-squared distribution was first introduced by the English statistician **Karl Pearson** in 1900. It has since become a fundamental concept in statistics and is extensively used in various statistical analyses.
***

### Important Properties:

<span style="font-size: 11pt; color: steelblue; font-weight:bold">Probability Density Function (PDF)</span>: 

$$f(x;k) = \frac{1}{2^{k/2} \Gamma(k/2)} \cdot x^{(k/2)-1} \cdot e^{-x/2}$$
where:
- $x$ is the random variable (non-negative real numbers).
- $k$ is the degrees of freedom (a positive integer).
- $\Gamma(\cdot)$ represents the gamma function.

<span style="font-size: 11pt; color: steelblue; font-weight:bold">Mean</span>: $$k$$

<span style="font-size: 11pt; color: steelblue; font-weight:bold">Median</span>: $$\approx k{\bigg (}1-{\frac {2}{9k}}{\bigg )}^{3}$$

<span style="font-size: 11pt; color: steelblue; font-weight:bold">Variance</span>: $$2k$$

<span style="font-size: 11pt; color: steelblue; font-weight:bold">Skewness</span>: $$\displaystyle {\sqrt {8/k}}$$

<span style="font-size: 11pt; color: steelblue; font-weight:bold">Kurtosis</span>: $${\displaystyle {\frac {12}{k}}}$$

### Applications:
The chi-squared distribution has several important use cases in statistics, including:
- Hypothesis testing: It is used to test the independence of variables or goodness-of-fit tests.
- Confidence intervals: It is used to construct confidence intervals for population variances.
- Sample variance distribution: It helps in understanding the distribution of sample variances of normally distributed populations.

### To Note:
- The sum of squares of independent standard normal random variables follows a chi-squared distribution.
- The chi-squared distribution is positively skewed and asymmetric.
- The chi-squared distribution is only defined for non-negative values, as it is based on the sum of squares.

The chi-squared distribution is a fundamental statistical tool that enables researchers and analysts to make inferences about population variances, test for independence in categorical data, and perform goodness-of-fit tests. It plays a crucial role in various statistical analyses, making it an essential concept for anyone working with data and statistics.

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings("ignore", category=RuntimeWarning)


# Set the degrees of freedom
dof = [1, 2, 3, 4, 5, 10]

# Plot the theoretical chi-squared PDFs
data = np.linspace(0, 30, 1000)

# Define colors and labels for the plots
colors = ['aqua', 'gold', 'springgreen', 'tomato', 'purple', 'brown']
labels = [f'Degrees of freedom: {d}' for d in dof]

# Set the figure size
plt.figure(figsize=(10, 5))

for i, d in enumerate(dof):
    
    # Get samples from chisquare distribution with dof = d
    samples = np.random.chisquare(d, size=1000)
    
    # Plot probability density functions for different 'k' values
    chi_squared_pdf = (1 / (2**(d/2) * np.math.gamma(d/2))) * data**(d/2 - 1) * np.exp(-data/2)
    plt.plot(data, chi_squared_pdf, color=colors[i], lw=2, label=labels[i])

plt.ylim(0, 0.5)
plt.xlim(0, 25)
plt.xlabel('Values')
plt.ylabel('Probability Density')
plt.title('Chi-Squared Distribution')
plt.grid(True, ls='dashed', alpha=0.3)
plt.legend()

plt.show()
```
