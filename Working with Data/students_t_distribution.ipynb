## <span style="font-size: 18pt; color: goldenrod; font-weight:bold">Student's t–distribution</span>

### Overview:
<u>The Student's t-distribution is a probability distribution that is used to model data when the sample size is small or when the population variance is unknown</u>. It is also known simply as the <span style="font-size: 11pt; color: goldenrod; font-weight:bold">t-distribution</span>. The t-distribution is similar to the standard normal (Gaussian) distribution but has heavier tails, which makes it more appropriate for small sample sizes or when the population variance is not known.

<span style="font-size: 11pt; color: mediumseagreen; font-weight:bold">The t-distribution is widely used in **inferential statistics**.</span>

**NOTE:** When working with small sample sizes (up to 30) or when the population variance is unknown, t-distribution provides a more accurate representation of uncertainty, compared to standard normal distribution.

### History:
The t-distribution was introduced by the English statistician **William Sealy Gosset** in **1908**. He used the pseudonym "Student" for publishing his work due to his employer's policy on not allowing employees to publish under their own names. Gosset worked for the Guinness brewery, and his work was related to quality control in brewing.

### Important Properties:

<span style="font-size: 11pt; color: steelblue; font-weight:bold">Probability Density Function (PDF)</span>:  

The probability density function of the Student's t-distribution is given by the following formula:

$$ \large f(x|\nu) = \frac{\Gamma\left(\frac{\nu+1}{2}\right)}{\sqrt{\nu\pi}\Gamma\left(\frac{\nu}{2}\right)} \left(1+\frac{x^2}{\nu}\right)^{-\frac{\nu+1}{2}} $$

where:
- $ x $ is the random variable.
- $ \nu $ (nu) is the degrees of freedom parameter.
- $ \Gamma $ is the gamma function.


<span style="font-size: 11pt; color: steelblue; font-weight:bold">Cumulative Distribution Function (CDF)</span>: 


$$\large F(x|\nu) = \frac{1}{2} + \frac{1}{2}\Gamma\left(\frac{\nu+1}{2}, \frac{x^2}{2}\right)\frac{1}{\Gamma(\nu/2)}$$

Where:
- Γ(a, x) represents the upper incomplete gamma function, defined as: $$\Gamma(a, x) = \int_{x}^{\infty} t^{a-1} e^{-t} dt$$
- Γ(a) represents the gamma function, defined as: $$\Gamma(a) = \int_{0}^{\infty} t^{a-1} e^{-t} dt$$


The degrees of freedom parameter $ \nu $ determines the shape of the t-distribution. <span style="font-size: 11pt; color: mediumseagreen; font-weight:bold">The larger the degrees of freedom, the more closely the t-distribution resembles the standard normal distribution.</span>

**NOTE:** Usually for a sample size $n$ we have $n-1$ degrees of freedom

### Application:
The t-distribution is commonly used in <u>hypothesis testing and confidence interval estimation</u> when dealing with *small sample sizes*. 

Some common use cases include:
- Comparing means of two samples with small sample sizes.
- Estimating confidence intervals for population parameters (e.g., mean or difference in means) when the population standard deviation is unknown.
- Conducting t-tests to assess the significance of a difference between two sample means.

The t-distribution finds application in various fields, including but not limited to:
- **Medical research**, when sample sizes are limited due to ethical or practical constraints.
- **Social sciences**, for conducting experiments or surveys with a small number of participants.
- **Quality control and manufacturing**, where a limited number of product samples are tested.

The Student's t-distribution is a fundamental concept in statistics and plays a crucial role in hypothesis testing and confidence interval estimation. Its ability to handle small sample sizes and unknown population standard deviations makes it an essential tool for making inferences about population parameters based on sample data.

#### Loading libraries

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import t
```

#### Compute and plot Student's t-distribution

```python
# Define the range of x values
data = np.linspace(-5, 5, 1000)

# Define the degrees of freedom for the t-distribution
degrees_of_freedom = [1, 2, 5, 20]
colors = ['aqua', 'gold', 'springgreen', 'tomato']

plt.figure(figsize=(8,4))
# Plot the PDF for each degree of freedom
for dof, col in zip(degrees_of_freedom, colors):
    # Create a t-distribution object with the given degrees of freedom
    t_dist = t(dof)
    
    # Calculate the PDF values for the given data range
    pdf_values = t_dist.pdf(data)
    
    # Plot the PDF curve
    plt.plot(data, pdf_values, color=col, label=f'degrees of freedom: {dof}')

# Add labels and legend
plt.xlabel('Values')
plt.ylabel('Probability Density')
plt.title("Student's t-distribution PDF")
plt.legend()

# Show the plot
plt.grid(True, ls='dashed', alpha=0.5)
plt.show()
```
