## <span style="font-size: 18pt; color: goldenrod; font-weight:bold">Gaussian (Normal) Destribution</u></span>
#### Overview

The <span style="font-size: 11pt; color: goldenrod; font-weight:bold">Normal Destribution</u></span>, also known as the Gaussian distribution, **is one of the most important and widely used probability distributions in statistics**. It is a <u>continuous probability distribution</u> that describes how a particular numerical variable is likely to be distributed. The distribution is <span style="font-size: 11pt; color: steelblue; font-weight:bold">symmetric and bell-shaped</span>, with the highest frequency of occurrences near the mean and decreasing probabilities as you move away from the mean.

<u>The Normal distribution is characterized by two parameters</u>: the mean $(\mu)$ and the standard deviation $(\sigma)$. The mean represents the center of the distribution, while the standard deviation controls the spread or dispersion of the data points around the mean.

***
The Normal distribution was first introduced by **Carl Friedrich Gauss**, a German mathematician and physicist, which is why it is sometimes referred to as the Gaussian distribution in his honor.
***

#### Important Properties:

- <span style="font-size: 11pt; color: steelblue; font-weight:bold">Probability Density Function (PDF)</span>:

$$f(x) = \frac{1}{\sigma \sqrt{2\pi}} \, e^{-\frac{(x-\mu)^2}{2\sigma^2}}$$

where:
- $f(x)$ is the probability density function at a point \(x\).
- $\mu$ is the mean of the distribution.
- $\sigma$ is the standard deviation of the distribution.
- $\pi$ is the mathematical constant pi (approximately 3.14159).
- $e$ is the mathematical constant Euler's number (approximately 2.71828).

Some other important statistical properties of the Normal distribution include:

1. <span style="font-size: 11pt; color: steelblue; font-weight:bold">Symmetry</span>: The Normal distribution is symmetric around its mean, which means that the probabilities of getting values above the mean are equal to the probabilities of getting values below the mean.

2. <span style="font-size: 11pt; color: steelblue; font-weight:bold">Mean and Median</span>: The mean and median of a Normal distribution are equal and located at the center of the distribution.

3. <span style="font-size: 11pt; color: steelblue; font-weight:bold">Empirical Rule (68-95-99.7 rule)</span>: For a Normal distribution, approximately $68%$ of the data falls within one standard deviation of the mean $(\mu \pm \sigma$), about $95%$ within two standard deviations $(\mu \pm 2\sigma)$, and about $99.7%$ within three standard deviations $(\mu \pm 3\sigma)$.

## Applications of Normal distribution:

The Normal distribution is widely used in various fields due to its versatility and importance. Some of its applications include:

1. **Inferential Statistics**: The Normal distribution serves as a fundamental assumption for many statistical tests and techniques. When certain conditions are met, it allows researchers to make inferences about populations based on sample data.

2. **Quality Control**: In manufacturing and production processes, the Normal distribution is often used to model the distribution of measurements and to set quality control limits.

3. **Financial Modeling**: In finance and economics, the Normal distribution is used to model asset returns, stock prices, and other financial variables.

4. **Natural Phenomena**: Many natural phenomena, such as the height of individuals in a population or errors in measurements, tend to follow a Normal distribution.

5. **Machine Learning**: In various machine learning algorithms, assumptions about the distribution of data are often made, and the Normal distribution is frequently used for modeling errors or noise in the data.

6. **Social Sciences**: The Normal distribution is used in social sciences to model human traits, IQ scores, and many other psychological and sociological variables.


```python
import numpy as np
import matplotlib.pyplot as plt
STD = '\u03C3'
MU = '\u03BC'

# Parameters for the normal distribution
mean = 0
standard_deviation = 1
sample_size = 10000

# Generate random samples from the normal distribution
samples = np.random.normal(mean, standard_deviation, sample_size)

plt.figure(figsize=(8,4))
# Create the histogram
plt.hist(samples, bins=50, density=True, color='steelblue', alpha=0.5)

# Calculate the range of x values to plot the theoretical PDF curve
x_values = np.linspace(mean - 4 * standard_deviation, mean + 4 * standard_deviation, 1000)

# Calculate the corresponding y values for the normal distribution's PDF
pdf_values = 1 / (standard_deviation * np.sqrt(2 * np.pi)) * np.exp(-(x_values - mean) ** 2 / (2 * standard_deviation ** 2))

# Plot the theoretical PDF curve as a line
plt.plot(x_values, pdf_values, color='seagreen', linewidth=2, label='PDF')

# Set plot labels and title
plt.xlabel('Values')
plt.ylabel('Probability Density')
plt.title(f'Histogram of a Normal Distribution ({MU} = {mean}, {STD}={standard_deviation})')
plt.legend()
plt.grid(True, ls='dashed', alpha=0.25)

# Show the plot
plt.show()
```
