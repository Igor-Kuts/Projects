## <span style="font-size: 24pt; color: goldenrod; font-weight:bold">Geometric Distribution</span>

### Overview:
The Geometric Distribution is a <span style="font-size: 11pt; color: steelblue; font-weight:bold">discrete probability distribution</span> that models the <u>number of trials needed to achieve the first success in a series of independent Bernoulli trials</u>. In each trial, there are only two possible outcomes: success (with probability $p$) or failure (with probability $1 - p$).

This distribution is particularly useful in probability and statistics, especially when dealing with scenarios where we want to know how many trials are needed before a success occurs.

The geometric distribution is a specific case of the negative binomial distribution, focusing on counting the number of failures before the first success is achieved.

### Important Properties:

<span style="font-size: 11pt; color: steelblue; font-weight:bold">Probability Mass Function (PMF)</span>:  

$$P(X=x)=(1−p)^{x−1} \cdot p$$

Where:
- $P(X=x)$ is the probability that the first success occurs on the $x$-th trial.
- $p$ is the probability of success on a single trial; $(0 < p < 1)$.
- $x$ is the number of trials before the first success; $x = 1,2 \dots n$

<span style="font-size: 11pt; color: steelblue; font-weight:bold">Cumulative Distribution Function (CDF)</span>:  

$$F(x) = P(X \leq x) = 1 - (1 - p)^x $$

<span style="font-size: 11pt; color: steelblue; font-weight:bold">Expected Value (Mean)</span>: $$\mu=\frac{1}{p}$$

<span style="font-size: 11pt; color: steelblue; font-weight:bold">Variance</span>: $$\sigma^2 = \frac{1-p}{p^2}$$

The geometric distribution is <span style="font-size: 11pt; color: steelblue; font-weight:bold">always right-skewed</span>, indicating that there is a higher probability of having a smaller number of trials before the first success. Additionally, the distribution is <span style="font-size: 11pt; color: steelblue; font-weight:bold">not upper-bounded</span>, meaning that the number of trials required for success can <u>*theoretically*</u> be infinite.

```python
import matplotlib.pyplot as plt
import numpy as np

def geometric_distribution_pmf(p, data):
    return (1 - p) ** (data - 1) * p

# Probability of success in independent Bernoulli trials
p_success = 1 / 6
data = np.arange(1, 16, 1)

# Calculate the probability of success on different number of Bernoulli trials.
pmf = geometric_distribution_pmf(p_success, data)

plt.figure(figsize=(8, 4))
bars = plt.bar(data, pmf)

# Add y-values atop each bar
for bar, value in zip(bars, pmf):
    plt.text(bar.get_x() + bar.get_width() / 2, bar.get_height() + 0.004,
             f"{value:.3f}", size=9, color='gray', style='normal',
             font='times new roman', ha='center', va='bottom')

plt.xticks(range(1, 16))

# Remove top and right frames
plt.gca().spines['top'].set_visible(False)
plt.gca().spines['right'].set_visible(False)

plt.xlabel('Number of Bernoulli trials before Success', size=10)
plt.ylabel('Probability Density', size=10)
plt.title('Geometric Distribution', size=16)

plt.show()
```
