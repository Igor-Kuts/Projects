## <span style="font-size: 18pt; color: goldenrod; font-weight:bold">Poisson Distribution</u></span>
#### Overview: 
The <span style="font-size: 11pt; color: goldenrod; font-weight:bold">Poisson Distribution</span><span style="font-size: 11pt; color: steelblue; font-weight:bold"> models the number of events that occur in a fixed interval of time or space</span>. It is often used for rare events with an average rate of occurrence $\lambda$.

**NOTE**: To apply the Poisson distribution, observed events must be occuring randomly and independently. Then $X$, the number of events in a fixed unit of time, has a Poisson distribution.

#### Important Properties:
- <span style="font-size: 11pt; color: steelblue; font-weight:bold">Probability Mass Function (PMF)</span>: $$P(X = x) = \frac{\lambda^x \cdot e^{-\lambda}}{x!}$$ where $x$ is the number of events observed.

- <span style="font-size: 11pt; color: steelblue; font-weight:bold">Mean</span>: $$\mu = \lambda $$
- <span style="font-size: 11pt; color: steelblue; font-weight:bold">Variance</span>: $$\sigma^2 = \lambda $$

#### Problem Example:
One nanogram of Plutonium-239 will have an average of 2.3 radioactive decays per second, and the number of decays will follow a Poisson distribution.  
What is the probability that in a 2 second period there are exactly 3 radioactive decays?

Solution:  
Let $X$ represent the number of decays in a 2 second period.  
Lambda is the mean number of decays <u> in that period</u>: $\lambda = 2.3 \cdot 2 = 4.6$  

$P(x = 3) = (\lambda^x \cdot e^{-\lambda}) \div x! = 0.163$

The probability of getting <u>exactly</u> 3 radioactive decays in 2 second period is 16.3%

```python
import numpy as np
import matplotlib.pyplot as plt
LAMBDA = '\u03BB'

lambda_ = 4.6  # Average rate or number of events in the interval

x_values = np.arange(0, 16)  # Possible number of events, ranging from 0 to 10
pmf_values = np.array([np.exp(-lambda_) * (lambda_**x) / np.math.factorial(x) for x in x_values])

plt.figure(figsize=(8,4))
plt.bar(x_values, pmf_values, align='center', alpha=0.5)
plt.xticks(x_values)
plt.xlabel('Number of Events (x)')
plt.ylabel('Probability')
plt.title(f'Probability Mass Function of the Poisson Distribution ({LAMBDA} = {lambda_})')
plt.grid(True, ls='dashed', alpha=0.25)

# Add y-values (probabilities) above bar number 3
for k, prob in zip(x_values, pmf_values):
    if k == 3:
        plt.text(k, prob, f'{prob*100:.1f}', ha='center', va='bottom')
plt.show()



```
