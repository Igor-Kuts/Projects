## <span style="font-size: 18pt; color: goldenrod; font-weight:bold">Binomial Distribution</u></span>
#### Overview:
The <span style="font-size: 11pt; color: goldenrod; font-weight:bold">Binomial Distribution</span> is an **extremely important** discreete probability destribution which has many applications. <span style="font-size: 11pt; color: steelblue; font-weight:bold">It describes the number of successes in a fixed number of independent Bernoulli trials</span>. It's used when <u>there are two possible outcomes with a **fixed probability** of success $p$ in each trial</u>.

#### Problem Example:
A fair, six–sided die is rolled 3 times (independent trials).  
What is the probability a 5 <u> comes up exactly twice</u>?

Success: Rolling a 5  
Failure: Rolling anything but a 5

Let $X$ represent the number of fives in 3 rolls.  
$X$ has a binomial distribution with $n = 3$ and $p = 1/6$

#### Important Properties:
- <span style="font-size: 11pt; color: steelblue; font-weight:bold">Probability Mass Function (PMF)</span>: $$ P(X = x) = \binom{n}{x} \cdot p^x \cdot (1 - p)^{n - x} $$ where $n$ is the number of trials, $k$ is the number of successes, $\binom{n}{x}$ is the binomial coefficient (*the number of ways to choose $k$ successes from $n$ trials – number of orderings*) and $p^x \cdot (1 - p)^{n - x}$ is the <u>probability of one specific ordering of successes and failures</u>.

- <span style="font-size: 11pt; color: steelblue; font-weight:bold">Mean</span>: $$\mu =  n \cdot p$$
- <span style="font-size: 11pt; color: steelblue; font-weight:bold">Variance</span>: $$\sigma^2 = n \cdot p \cdot (1 - p) $$


```python
import numpy as np
import matplotlib.pyplot as plt

n = 10    # Number of trials
p = 0.5   # Probability of success in each trial

x_values = np.arange(0, n+1)  # Possible number of successes, ranging from 0 to x
pmf_values = np.array([np.math.comb(n, k) * p**k * (1-p)**(n-k) for k in x_values])

plt.figure(figsize=(8,4))
plt.bar(x_values, pmf_values, align='center', alpha=0.5)

plt.xticks(x_values)
plt.xlabel('Number of Successes (k)')
plt.ylabel('Probability')
plt.title(f'Probability Mass Function of the Binomial Distribution (p = {p})')
plt.grid(True, ls='dashed', alpha=0.25)

plt.show()
```
