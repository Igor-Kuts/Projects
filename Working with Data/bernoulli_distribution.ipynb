## <span style="font-size: 18pt; color: goldenrod; font-weight:bold">Bernoulli Distribution</u></span>
#### Overview:
The <span style="font-size: 11pt; color: goldenrod; font-weight:bold">Bernoulli Distribution</u></span> is a <u>discreete probability distribution</u>. It models a binary outcome / <span style="font-size: 11pt; color: steelblue; font-weight:bold">trial</span>, where an<span style="font-size: 11pt; color: steelblue; font-weight:bold"> event can have only two possible outcomes</span>, usually labeled as "*success*" and "*failure*" (or $1$ and $0$). It is characterized by a single parameter, $p$, <u>representing the probability of success</u>.

**NOTE**: Some other common discrete probability distributions are built on the assumption of independent Bernoulli trials. (Geometric, binomial, negative binomial)

#### Important Properties:
- <span style="font-size: 11pt; color: steelblue; font-weight:bold">Probability Mass Function (PMF)</span>: 
$$ P(X = x) = p^{x} \cdot (1 - p)^{1 - x} \\
   \cdots                                 \\
   P(X=1) = p^1 \cdot (1 - p)^{1 - 1}=p   \\
   P(X=0) = p^0 \cdot(1 - p)^{1 - 0}= 1- p$$

- <span style="font-size: 11pt; color: steelblue; font-weight:bold">Mean</span> $$ \mu = p$$
- <span style="font-size: 11pt; color: steelblue; font-weight:bold">Variance</span> $$\sigma^2 = p \cdot (1 - p) $$

```python
import matplotlib.pyplot as plt

# Define the probability of success (getting a 1) in a Bernoulli trial
p = 0.3

# Define the outcomes and their corresponding probabilities
outcomes = [0, 1]
probabilities = [1 - p, p]

# Create the bar plot for the PMF
plt.figure(figsize=(8,4))
plt.bar(outcomes, probabilities, align='center', alpha=0.5)

plt.xticks(outcomes)
plt.xlabel('Outcomes')
plt.ylabel('Probability')
plt.title(f'Probability Mass Function of Bernoulli Distribution (p = {p})')
plt.grid(True, ls='dashed', alpha=0.25)

plt.show()
```
