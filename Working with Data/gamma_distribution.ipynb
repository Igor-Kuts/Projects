## <span style="font-size: 18pt; color: goldenrod; font-weight:bold">Gamma Distribution</u></span>

The <span style="font-size: 11pt; color: goldenrod; font-weight:bold">Gamma Distribution</u></span> is a <u>continuous probability distribution</u> that is commonly used in statistics and probability theory to model *positive-valued random variables*. It is widely used in various fields, such as physics, engineering, finance, and life sciences, where variables are constrained to be positive.

- Gamma distribution is a generalization of Chi-Square distribution.
- Gamma distribution is a continuous and univariate distribution.  
- Gamma random variable can take any positive real value: $[0,\infty)$  
- Gamma distribution <span style="font-size: 11pt; color: steelblue; font-weight:bold">has several equivalent parametrizations</span> and the first thing we need to do when working with Gamma distribution is to <u>check which parametrization is being used</u>. 

#### How it is obtained:
  
Gamma distribution is obtained as a Sum of $n$ <u>independent normals</u> with (mean 0, variance $\sigma^2$)   

It has two parameters: 
- number of variables in the sum $n$ 
- and their variance $\sigma^2$ 

<span style="font-size: 11pt; color: steelblue; font-weight:bold">In general, the sum of independent squared normal variables that have zero mean and arbitrary variance has a Gamma distribution.</span>

#### Probability Density function (PDF):
This parametrization is particularly convinient and often used in Bayesian applications, because it simplifies calculations.

$$fx(x) = \begin{cases} \textbf{c}x^{n/2-1}exp(-\frac{n}{h}\frac{1}{2}x) & \text{if } x \in R_x \\
0 & \text{if } x  \notin R_x \end{cases}$$

where $\textbf{h}$ is the mean, $\textbf{n}$ is degrees of freedom and $\textbf{c}$ is a constant:
$$ \textbf{c} = \frac{(n/h)^{n/2}}{2^{n/2}\Gamma(n/2)} $$

and $\Gamma( )$ is the **Gamma function**

#### Mean and Variance:

The expected value (mean) and variance of the gamma distribution are given by:

- Mean: $\mathbb{E}[X] = h$
- Variance: $Var[X] = 2\frac{h^2}{n}$

#### Moment generating function:

The moment generating function of a Gamma random variable $x$ is defined for any $t < \frac{n}{2h}$:
$$ M_x(t) = (1 - \frac{2h}{n}t)^{-n/2}$$

#### Characteristic function:

The characteristic function of a Gamma random variable $x$ is:
$$\phi x(t) = (1 - \frac{2h}{n}it)^{-n/2}$$

#### Distribution function:

The distribution function of a Gamma random variable is:
$$F_x(x) = \frac{\gamma(n/2, nx/2h)}{\Gamma(n/2)}$$

where the function 
$$\gamma(n/2, nx/2h) = \int_{-\infty}^{nx/2h}s^{n/2-1}exp(-s)ds$$

is called lower incomplete Gamma function and is usually evaluated using specialized computer algorithms.

#### Use in modeling:
The shape and scale parameters allow the Gamma distribution to take on a wide range of shapes, making it quite flexible in modeling various data distributions. It is often used in combination with other statistical techniques, such as Maximum Likelihood Estimation, to fit the distribution to empirical data and make inferences about the underlying population.

## Probability Density Function of Gamma Distribution with visualization

1. The first one (blue) is the PDF of a Gamma random variable with $n=6$ degrees of freedom and mean $h=3$.
2. The second one (orange) is obtained by setting $n=8$ and $h=3$

Because of $h=3$ in both cases, the two distributions have the same mean.
However, by increasing $n$ from $6$ to $8$, the shape of the distribution changes. 

<span style="font-size: 11pt; color: mediumseagreen; font-weight:bold">The more we increase the degrees of freedom, the more the PDF resembles that of a Normal Distribution.</u></span>

```python
import math
import numpy as np
import matplotlib.pyplot as plt

# Calculate the gamma function for a given value z
def gamma_function(z):
    return math.gamma(z)

# Function to calculate the constant 'c' in the gamma PDF formula
def c(n, h):
    enom = (n / h) ** (n / 2)
    denom = 2 ** (n / 2) * gamma_function(n / 2)
    return enom / denom

# Function to calculate the gamma PDF for a given 'n', 'h', and 'x'
def f_x(n, h, x):
    return c(n, h) * x ** (n / 2 - 1) * np.exp(-(n / h) * (1 / 2) * x)

# Generate x values for the plot
x = np.linspace(0, 20, 1000)

# Define degrees of freedom (n) and expected value (h)
n_1, n_2 = 6, 8
h = 3
std = np.std(x)

# Calculate the gamma PDFs for the two sets of degrees of freedom
gamma_pdf_1 = f_x(n_1, h, x)
gamma_pdf_2 = f_x(n_2, h, x)

# Plot the gamma PDFs on the same figure
plt.figure(figsize=(8, 4))
plt.plot(x, gamma_pdf_1, label=f'PDF with {n_1} degrees of freedom')
plt.plot(x, gamma_pdf_2, label=f'PDF with {n_2} degrees of freedom')

# Plot Customization
plt.xlabel('Values')
plt.ylabel('Probability Density')
plt.title('Probability Density Function of Gamma Distribution')
plt.legend()
plt.grid(True, ls='dashed', alpha=0.5)

plt.show()
```
