## <span style="font-size: 18pt; color: goldenrod; font-weight:bold">Exponential Distribution</u></span>

The <span style="font-size: 11pt; color: goldenrod; font-weight:bold">Exponential Distribution</u></span> is a continuous distribution which describes distance (or length, area, volume, time) between events in a Poisson process, where $x â€” \textbf{Continuous Interval between two events}$. In a Poisson process, discrete events occur <u>randomly and independently</u> over time in a continuous interval, but the <span style="font-size: 11pt; color: steelblue; font-weight:bold">time between consecutive events in Poisson Process follows an exponential distribution</u></span>.   

<span style="font-size: 11pt; color: goldenrod; font-weight:bold">Exponential Distribution</u></span> is often used to model the time until the next event occurs in various real-world scenarios, such as arrival times, waiting times, or time until failure, or natural disasters.

#### Probability Density Function (PDF):

The probability density function of the exponential distribution is given by:

$$f(x;\lambda) = \lambda e^{-\lambda x}$$

Often, we are interested in the cumulative exponential distribution function $P[0:X]$:

$$F(x)=1 - e^{-\lambda x}$$


where:
- $x$ is the time variable for which we want to find the probability of an event occurring.
- $\lambda$ is the rate parameter (also known as the rate of occurrence or the average number of events per unit of time).

#### Mean and Variance:

The mean (or expected value) and variance of the exponential distribution are given by:

- Mean: $\mu = \frac{1}{\lambda}$
- Variance: $\sigma^2 = \frac{1}{\lambda^2}$

#### Problem Example:

Suppose the time between successive radioactive decays in a sample of Plutonium-239 follows an exponential distribution with a decay rate of $\lambda = 2.3$ decays per second. What is the probability that the time between two consecutive decays is at most 0.5 seconds?

Solution:

We are interested in finding $P(X \leq 0.5)$, where $X$ represents the time between consecutive decays.

Using the exponential PDF, we have:

$$P(X \leq 0.5) = \int_{0}^{0.5} \lambda e^{-\lambda x} \, dx = \left[-e^{-\lambda x}\right]_{0}^{0.5} = -e^{-\lambda \cdot 0.5} + e^{-\lambda \cdot 0} = 1 - e^{-1.15} \approx 0.6845$$

The probability that the time between two consecutive decays is at most 0.5 seconds is approximately 68.45%.



## Example with solution and visualization
#### Problem statement

In the last 40 years, there have been 200 earthquakes.

1. What is the probability that there will be at least 3 months until the next earthquake?
2. What is the probability that there will less than 3 months until the next earthquake?

- We have to find the probability $P[X \geq 3]$
- We have to find the probability $P[X < 3]$

```python
import matplotlib.pyplot as plt
import numpy as np

# Given data
num_earthquakes = 200
num_years = 40
num_months = num_years * 12
average_frequency = num_earthquakes / num_months

# Time between consecutive earthquakes follows an exponential distribution
time_between_earthquakes = 1 / average_frequency

# Calculate probabilities using the exponential distribution formula
probability_at_least_3 = np.exp(-3 * average_frequency)
probability_less_than_3 = 1 - probability_at_least_3

# Visualize the exponential distribution
x = np.linspace(0, 10, 10_000)
exponential_pdf = average_frequency * np.exp(-average_frequency * x)

plt.figure(figsize=(8,4))
plt.plot(x, exponential_pdf)
plt.xticks(np.arange(0,10,1))
plt.xlabel('Time between Earthquakes (months)')
plt.ylabel('Probability Density')
plt.title('Exponential Distribution of Time Between Earthquakes')
plt.grid(True, ls='dashed', alpha=0.5)

# Fill the area under the curve corresponding to the probability P(X >= 3)
x_fill_at_least_3 = np.linspace(3, 10, 1000)
plt.fill_between(x_fill_at_least_3, average_frequency * np.exp(-average_frequency * x_fill_at_least_3), color='mediumseagreen', alpha=0.6)

# Fill the area under the curve corresponding to the probability P(X < 3)
x_fill_less_than_3 = np.linspace(0, 3, 1000)
plt.fill_between(x_fill_less_than_3, average_frequency * np.exp(-average_frequency * x_fill_less_than_3), color='orange', alpha=0.6)

plt.legend(["Exponential PDF", "P(X >= 3)", "P(X < 3)"])
plt.show()

# Print the probabilities
print(f"Probability of at least 3 months until the next earthquake: {probability_at_least_3:.4f}")
print(f"Probability of less than 3 months until the next earthquake: {probability_less_than_3:.4f}")
```
