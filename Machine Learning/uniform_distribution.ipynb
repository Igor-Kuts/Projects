## <span style="font-size: 18pt; color: goldenrod; font-weight:bold">Uniform Distribution</u></span>
#### Overview:
In a <span style="font-size: 11pt; color: goldenrod; font-weight:bold">Uniform Distribution</span>, all outcomes within a given range have an equal probability of occurring. It is <u>characterized by a flat, rectangular shape on a graph</u>. 

**NOTE**: In Uniform Distribution both the mean and the variance are uninterpretable and posses no predictive power.  
**NOTE**: The probability density function (PDF) for a Continuous Uniform Distribution is constant over the interval $[a, b]$, and for a Discrete Uniform Distribution, the probability mass function (PMF) assigns equal probabilities to each outcome in a finite set.

#### Important Properties:
- <span style="font-size: 11pt; color: steelblue; font-weight:bold">Probability Density Function (PDF)</span>:
  - Continuous Uniform: $$ f(x) = \begin{cases} \frac{1}{b-a} & \text{for } a \leq x \leq d, \\ 0 & \text{elsewhere}\end{cases} $$
  - Discrete Uniform: $$ P(X = x) = \frac{1}{b - a + 1} \text{ for } x \in \{a, a+1, \dots, b\}$$

- <span style="font-size: 11pt; color: steelblue; font-weight:bold">Mean</span>: $$\mu =\frac{a + b}{2}$$
- <span style="font-size: 11pt; color: steelblue; font-weight:bold">Variance</span>:
    - Continuous Uniform: $$\sigma^2 = \frac{(b - a)^2}{12}$$
    - Discrete Uniform: $$\sigma^2 = \frac{(b - a + 1)^2}{12}$$

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the range of the continuous uniform distribution
lower_bound = 1
upper_bound = 11

# Compute the constant PDF value for the uniform distribution
pdf_value = 1 / (upper_bound - lower_bound)

# Create the plot using Matplotlib's plot function
plt.figure(figsize=(8,4))
plt.plot([lower_bound, upper_bound], [pdf_value, pdf_value], linewidth=2, alpha=0.5)

# Fill the area between the two points with color to create a rectangle
plt.fill_between([lower_bound, upper_bound], [0, 0], [pdf_value, pdf_value], alpha=0.5)

plt.xticks(np.arange(0, 12, 1))
plt.xlim(0, 12)
plt.ylim(0, 0.15)
plt.xlabel('Values')
plt.ylabel('Probability Density')
plt.title('Probability Density Function of Continuous Uniform Distribution')
plt.grid(True, ls='dashed', alpha=0.25)

plt.show()
```
