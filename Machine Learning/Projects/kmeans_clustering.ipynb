# K-Means clustering algorithm
   

**K-Means**  is <u>one of the most popular</u> <span style="font-size: 11pt; color: steelblue; font-weight: bold">**Unsupervised Machine Learning Algorithms**</span> used for partitioning a dataset into distinct groups or clusters based on their similarity. The algorithm aims to minimize the *intra-cluster variance*, which helps in identifying natural groupings in the data.

## Overview
The K-Means algorithm works by iteratively assigning data points to clusters and updating the cluster centers until convergence or stoppage criteria is met. <u>It requires the number of clusters $(K)$ to be specified in advance</u>.

## History

The K-Means clustering algorithm was first introduced by **Stuart Lloyd** in 1957 as a method for pulse-code modulation in telecommunications. However, it was later independently rediscovered and popularized by **James MacQueen** in 1967 and **J. B. MacQueen** in 1967 and **J. A. Hartigan** and **M. A. Wong** in 1979.  

Since then, it has become one of the most widely used clustering algorithms in the field of Machine Learning and Data Mining.

## Algorithm:

1. **Initialization**: Randomly initialize K cluster centers in the feature space. Each cluster center represents the <span style="font-size: 11pt; color: steelblue; font-weight: bold">centroid of the points assigned to that cluster</span> .

2. **Assignment**: <u>For each data point</u>, <u>calculate its distance to each cluster</u> center using a distance metric such as <span style="font-size: 11pt; color: goldenrod; font-weight: bold">Euclidean distance</span> or others. <u>Assign the point to the cluster with the nearest cluster center</u>.

3. **Update**: Recalculate the centroids by taking the mean of all the data points assigned to each cluster. This step aims to adjust the cluster centers to better represent the data points within each cluster.

4. **Convergence Check**: Repeat the assignment and update steps until convergence or a stopping criterion is met. This criterion can be a maximum number of iterations, a threshold for the change in cluster centers, or when the assignments no longer change.

5. **Final Result**: The algorithm outputs the final cluster assignments, where each data point belongs to the cluster with the closest cluster center.

## Computation:

First we will define some mathematical notations used in the K-Means:

- Let $X = (x_1, x_2, \dots, x_n)$ be the dataset with $n$ data points, where each <u>data point $x_i$ is a $d$-dimensional vector</u>.
- Let $C = (c_1, c_2, \dots, c_n)$ be the <u>set of centroids</u>, where each centroid $c_i$ is also a $d$-dimensional vector.
- Let $A = (a_1, a_2, \dots, a_n)$ be the <u>set of cluster assignments</u>, where each assignment $a_i$ indicates the index of the centroid that $x_i$ is assigned to.

#### <span style="font-size: 11pt; color: steelblue; font-weight: bold">Initialization</span>: 
<u>Randomly</u> initialize centroids $C = \{c_1, c_2, ..., c_k\}$.

#### <span style="font-size: 11pt; color: steelblue; font-weight: bold">Assignment</span>:
For each data point $x_i$:
- Calculate the distance between $x_i$ (such as Euclidean) and each centroid $c_j$:

   $$d(x_i, c_j) = ||x_i - c_j||^2$$
   
- Assign $x_i$ to the centroid with the minimum distance:

   $$a_i = \arg \min_j (d(x_i, c_j))$$
#### <span style="font-size: 11pt; color: steelblue; font-weight: bold">Update</span>:
For each centroid $c_j$:
- Calculate the new centroid location by taking the mean of the data points assigned to that centroid:

   $$c_j = \frac{1}{|S_j|} \sum_{x \in S_j} x\text{, where }S_j = \{x_i | a_i = j\}$$

#### <span style="font-size: 11pt; color: steelblue; font-weight: bold">Convergence Check</span>:
Repeat the <span style="font-size: 11pt; color: steelblue; font-weight: bold">Assignment</span> and <span style="font-size: 11pt; color: steelblue; font-weight: bold">Update</span> steps until convergence or a stopping criterion is met.

#### <span style="font-size: 11pt; color: steelblue; font-weight: bold">Final Result</span>:

Output the cluster assignments $A = \{a_1, a_2, ..., a_n\}$.

## Pros and Cons
#### Pros:
- K-Means is relatively easy to understand and implement.
- It is <span style="font-size: 11pt; color: mediumseagreen; font-weight: bold">computationally efficient</span> and works well with large datasets.
- The algorithm scales well with the number of data points and clusters.
- It is <span style="font-size: 11pt; color: mediumseagreen; font-weight: bold">applicable to a wide range of domains</span> and <u>has been successfully used in various fields</u>.
- The results of K-Means clustering can be <span style="font-size: 11pt; color: mediumseagreen; font-weight: bold">interpreted and visualized easily</span> and <u>.

#### Cons:
- K-Means <span style="font-size: 11pt; color: orange; font-weight: bold">requires the number of clusters to be specified in advance</span>, which can be challenging in some cases.
- The algorithm is <span style="font-size: 11pt; color: orange; font-weight: bold">sensitive to the initial placement of cluster centers</span>, as it may converge to different solutions.
- It <u>assumes that clusters are spherical</u> and have <u>similar variances</u>, which might not be suitable for all types of data.
- K-Means <span style="font-size: 11pt; color: orange; font-weight: bold">can be influenced by outliers</span>, as they can significantly affect the cluster centers' positions.
- The algorithm does not handle categorical or missing data directly and may require preprocessing steps.

Our objective is to utilize the K-Means algorithm to partition the data into meaningful clusters. To achieve this, we will first create a class for implementing the K-Means algorithm. Following that, we will visualize the resulting clusters by plotting them.

````python
class KMeans:
    """
    K-Means clustering algorithm implementation.

    Parameters:
        n_clusters (int): Number of clusters (default: 8).
        max_iters (int): Maximum number of iterations (default: 10).
        plot_progress (bool): Flag indicating whether to plot progress (default: False).
    """

    def __init__(self, n_clusters=8, max_iters=10, plot_progress=False):
        self.n_clusters = n_clusters
        self.max_iters = max_iters
        self.plot_progress = plot_progress
        self.centroids = None
        self.idx = None

    def draw_line(self, p1, p2, linewidth=0.7):
        """
        Draw a line segment between two points.

        Parameters:
            p1 (tuple): First point coordinates (x, y).
            p2 (tuple): Second point coordinates (x, y).
            linewidth (float): Line width (default: 0.7).
        """
        plt.plot([p1[0], p2[0]], [p1[1], p2[1]], c='k', ls='dashed', linewidth=linewidth, alpha=0.5)

    def plot_data_points(self, X):
        """
        Plot data points with different colors based on cluster assignments.

        Parameters:
            X (ndarray): Data matrix, where each row is a single example.
        """
        cmap = ListedColormap(["tomato", "lawngreen", "dodgerblue", "black"])
        c = cmap(self.idx)
        plt.scatter(X[:, 0], X[:, 1], facecolors='none', edgecolors=c, linewidth=0.3, alpha=0.4)
        plt.xlabel('Feature 1')
        plt.ylabel('Feature 2')
        plt.grid(True, ls='dashed', alpha=0.5)
        
    def plot_progress_kMeans(self, X, i):
        """
        Plot the data points, centroids, and centroid movement lines.

        Parameters:
            X (ndarray): Data matrix, where each row is a single example.
            i (int): Current iteration number.
        """
        self.plot_data_points(X)
        plt.scatter(self.centroids[:, 0], self.centroids[:, 1], marker='o', facecolors='cyan', edgecolors='k', linewidths=1.5)
        for j in range(self.centroids.shape[0]):
            self.draw_line(self.centroids[j, :], self.previous_centroids[j, :])
        plt.title(f"Iteration number {i}")

    def fit(self, X):
        """
        Runs the K-Means algorithm on data matrix X.

        Parameters:
            X (ndarray): Data matrix, where each row is a single example.
        """
        m, n = X.shape
        self.centroids = X[np.random.choice(m, self.n_clusters, replace=False)]
        self.previous_centroids = self.centroids.copy()
        self.idx = np.zeros(m)

        if self.plot_progress:
            plt.figure(figsize=(8, 5))

        for i in range(self.max_iters):
            print(f"K-Means iteration {i}/{self.max_iters-1}")
            self.idx = self.find_closest_centroids(X)

            if self.plot_progress:
                self.plot_progress_kMeans(X, i)
                self.previous_centroids = self.centroids.copy()

            self.centroids = self.compute_centroids(X)

        if self.plot_progress:
            plt.show()

    def compute_centroids(self, X):
        """
        Returns the new centroids by computing the means of the data points assigned to each centroid.

        Parameters:
            X (ndarray): Data matrix, where each row is a single example.

        Returns:
            centroids (ndarray): Final centroid positions.
        """
        centroids = np.zeros((self.n_clusters, X.shape[1]))
        for k in range(self.n_clusters):
            points = X[self.idx == k]
            centroids[k] = np.mean(points, axis=0)
        return centroids

    def find_closest_centroids(self, X):
        """
        Computes the centroid memberships for every example.

        Parameters:
            X (ndarray): Data matrix, where each row is a single example.

        Returns:
            idx (ndarray): Cluster assignments for each example.
        """
        distances = np.zeros((X.shape[0], self.n_clusters))
        for j in range(self.n_clusters):
            distances[:, j] = np.linalg.norm(X - self.centroids[j], axis=1)
        return np.argmin(distances, axis=1)
````

After running the K-Means algorithm, <u>we can easily obtain the data points belonging to each cluster by accessing the cluster assignments generated by the algorithm</u>. In the context of the implemented `KMeans` class, we can access the cluster assignments through the `idx` attribute.

The `idx` attribute is an array of cluster assignments, where each element corresponds to the cluster index that a data point is assigned to. To obtain the data points for a specific cluster, we can simply filter the data based on the cluster assignment.

For example, let's say you have performed K-Means clustering on a dataset `X` and obtained the cluster assignments in the `idx` array. To get the data points for cluster `k`, you can do the following:

```python
data_cluster_k = X[KMeans.idx == k]
```

This will create a new array `data_cluster_k` that contains only the data points assigned to cluster `k`.
