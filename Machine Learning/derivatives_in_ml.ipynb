# Derivatives in Machine Learning

**Derivatives** play very important role in Machine Learning and Neural Networks. One of the most common place where we are encountering derivatives is the <span style="font-size: 11pt; color:orange; font-weight: bold">Backpropagation of Errors</span> in Neural Networks.  

During backward propagation of errors our goal is to minimize the $\text{Loss Function}$ of the model by calculating it's derivative with respect to $\text{weights}$ and $\text{biases}$:

1. First we calculate the derivative of the $\text{Loss Function}$ with respect to the model's outputs.
2. Next, we propagate this derivative backwards throught the layers of the Neural Network, applying the <span style="font-size: 11pt; color:tomato; font-weight: bold">Chain Rule</span> to compute the derivatives of the $\text{Loss Function}$ with respect to the $\text{weights}$ and $\text{biases}$ at each layer.

To calculate these derivatives, we will have to compute the derivative of the activation function used in each layer. This is because the derivative of the output of a layer depends on the derivative of it's activation function.
    
Different <span style="font-size: 11pt; color:steelblue; font-weight: bold">Activation Functions</span> have different derivatives, so it is essential to know how to compute them.

Next we will take a look at computation and graphs of the derivatives for three most common activation functions used in Neural Networks: 
- Sigmoid
- Rectified Linear Unit (ReLU)
- Hyperbolic Tangent (Tanh)


```python
# Importing libraries
import matplotlib.pyplot as plt
import numpy as np
```

### Sidmoid Derivative:
$$\sigma'(x) = \sigma(x) \cdot (1 - \sigma(x)) $$

```python
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

def sigmoid_derivative(x):
    return sigmoid(x) * (1 - sigmoid(x))

# Synthetic data
data = np.linspace(-5,5,100)

# Apply Sigmoid and Sigmoid derivative
data_sigmoid = sigmoid(data)
data_sigmoid_derivative = sigmoid_derivative(data)

# Create plot
plt.figure(figsize=(8,4))
plt.plot(np.arange(0,10,0.1), data_sigmoid, label=('Sigmoid'), lw=3)
plt.plot(np.arange(0,10 ,0.1), data_sigmoid_derivative, label=('Sigmoid Derivative'), lw=3)

# Customize plot
plt.xticks(np.arange(0,11))
plt.xlim(0,10)
plt.yticks(np.arange(0,1.1,0.1))
plt.ylim(0,1)
plt.grid(True, ls='dashed', alpha=0.5)
plt.legend()

plt.show()
```

### ReLU Derivative
$$ ReLU'(x) = 
\begin{cases}
    0  &\quad \text{if } x  \leq 0,\\
    1  & \quad \text{if } x > 0
  \end{cases}
  $$

```python
def relu(x):
    return np.maximum(0, x)

def relu_derivative(x):
    return (x > 0) * 1

# Synthetic data
data = np.linspace(-1,1,1000)

# Apply ReLU and ReLU derivative
data_relu = relu(data)
data_relu_derivative = relu_derivative(data)

# Create plot
plt.figure(figsize=(8,4))
plt.plot(np.arange(-5,5,0.01), data_relu, label=('ReLU'), lw=3)
plt.plot(np.arange(-5,5,0.01), data_relu_derivative, label=('ReLU Derivative'), lw=3)

# Customize plot
plt.gca().spines['left'].set_position('zero')
plt.gca().spines['bottom'].set_position('zero')
plt.gca().spines['right'].set_color('none')
plt.gca().spines['top'].set_color('none')
plt.xticks(np.arange(-5,6,1))
plt.yticks(np.arange(0,1.1,0.2))
plt.grid(True, ls='dashed', alpha=0.5)
plt.legend()

plt.show()
```

# Hyperbolic Tangent Derivative
$$\text{tanh'}(x) = 1 - \text{tanh}(x)^2$$ 

```python
def tanh_derivative(x):
    return 1 - np.tanh(x)**2

# Synthetic data
data = np.linspace(-5,5,1000)

# Apply TanH and TanH derivative
data_tanh = np.tanh(data)
data_tanh_derivative = tanh_derivative(data)

# Create plot
plt.figure(figsize=(8,4))
plt.plot(np.arange(-5,5,0.01), data_tanh, label=('TanH'), lw=3)
plt.plot(np.arange(-5,5,0.01), data_tanh_derivative, label=('TanH Derivative'), lw=3)

# Customize plot
plt.gca().spines['left'].set_position('zero')
plt.gca().spines['bottom'].set_position('zero')
plt.gca().spines['right'].set_color('none')
plt.gca().spines['top'].set_color('none')
plt.yticks(np.arange(-1,1.1,0.5))
plt.grid(True, ls='dashed', alpha=0.5)
plt.legend()

plt.show()
```
