# Linear Regression

**Linear Regression** is a <u>supervised learning algorithm</u> used for predicting a continuous target variable based on one or more input features. It assumes a linear relationship between the input features and the target variable.  

The goal of linear regression is to find the best-fit line that minimizes the difference between the predicted values and the actual values.

Linear regression is the basis for many other regression techniques, such as polynomial regression, ridge regression and lasso regression.  

Variants of linear regression, such as logistic regression, are used for classification tasks in machine learning.

## Formulas

In simple linear regression, which involves one input feature, the relationship between the input feature $x$ and the target variable $y$ is represented by the equation:

$$y = wx+b$$

Here, $b$ is the y-intercept (the value of y when x is 0), and $w$ is the slope of the line (the change in y for a unit change in x). The goal is to estimate the values of $b$ and $w$ that minimize the sum of squared differences between the predicted and actual values.

The estimation of the coefficients is typically done using the least squares method, which involves minimizing the sum of squared residuals between the predicted and actual values (**MSE**).

Optimization algorithms like **Gradient Descent**, **Stochastic Gradient Descent**, **Adaptive Moment Estimation**, and **Sophia** are commonly used to iteratively update the weights during training.

## History
Linear regression has a long history and has been extensively studied in statistics and machine learning. The method was first introduced by **Francis Galton** in the 19th century and later developed by **Sir Francis Ysidro Edgeworth** and **Karl Pearson**. Since then, it has become a fundamental tool in statistics and data analysis.


## Usage

Linear regression can be easily implemented by using the scikit-learn library:

```python
from sklearn.linear_model import LinearRegression

# Create a linear regression object
regression = LinearRegression()

# Fit the regression model to the training data
regression.fit(X_train, y_train)

# Predict the target variable for new data
y_pred = regression.predict(X_test)
```

In this code, `X_train` and `y_train` represent the training data, and `X_test` contains the input features for new data that we want to predict. The `fit()` method is used to train the linear regression model, and the `predict()` method is used to make predictions based on the trained model.

## Applications

Linear regression has numerous applications across various domains. Some common applications include:
- Predicting house / car / insurance, etc. prices based on factors such as location, size, and number of rooms.
- Forecasting sales based on advertising expenditure, historical data, and other relevant variables.
- Analyzing the relationship between independent and dependent variables in scientific research.
- Predicting a person's income based on education level, work experience, and other demographic factors.

## Pros and Cons

#### Pros of linear regression:
- **Simplicity**: Linear regression is easy to understand and implement.
- **Interpretability**: The coefficients in linear regression provide insights into the relationship between input features and the target variable.
- **Efficiency**: Linear regression can be <span style="font-size: 11pt; color: green; font-weight: bold">computationally efficient</span> and can handle large datasets.
- **Wide applicability**: Linear regression <span style="font-size: 11pt; color: green; font-weight: bold">can be applied to a wide range of problems</span>, from simple to complex.

#### Cons of linear regression:
- **Linearity assumption**: Linear regression assumes a linear relationship between the input features and the target variable. <span style="font-size: 11pt; color: orange; font-weight: bold">If the relationship is nonlinear, linear regression may not perform well</span>.
- **Sensitivity to outliers**: <span style="font-size: 11pt; color: orange; font-weight: bold">Linear regression is sensitive to outliers</span>, which can disproportionately affect the estimated coefficients.
- **Limited complexity**: <u>Linear regression is a linear model</u> and may not capture complex relationships present in the data.

# Example usage

#### Importing libraries

````python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
````

#### Creating synthetic data and training Linear Regression

````python
np.random.seed(40)

# Input feature (X) and corresponding target variable (y)
X_train = np.random.randint(low=0, high=42, size=(50,))
y_train = 2.34 * X_train

X_test = np.random.randint(low=0, high=42, size=(50,))
y_test = 2.31 * X_test

# Adding noise to y_train and y_test
mean = 2  # Mean of the noise
std_dev = 7  # Standard deviation of the noise
noise = np.random.normal(mean, std_dev, size=(50,))
y_train = y_train + noise
y_test = y_test + noise

# Creating LinearRegression object
model = LinearRegression()

# Fitting the model to the data
model.fit(X_train.reshape(-1, 1), y_train)

# Making predictions on unseen test data
y_pred = model.predict(X_test.reshape(-1, 1))

# Calculating mean squared error of the predictions
mse_lr = mean_squared_error(y_test, y_pred)
print('Mean Squared Error on test data:', mse_lr)

# Plotting the actual data points
plt.scatter(X_test, y_test, color='silver', label='Actual Data', s=15)

# Plotting the predicted values
plt.plot(X_test, y_pred, color='orange', label='Predicted Values', lw=1)

# Adding labels and title
plt.xlabel('Input Feature (X)')
plt.ylabel('Target Variable (y)')
plt.title('Linear Regression')
plt.grid(True, lw=0.3,  ls='dashed', alpha=0.5)

# Displaying the plot
plt.legend()
plt.show()
````
