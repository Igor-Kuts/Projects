# Coefficient of Determination (R2)

## Overview:

**Coefficient of Determination**, $R^2$ serves as a measure of correlation between two variables. $R^2$ is a statistical measure used to assess the goodness of fit of a regression model. In the realm of Machine Learning, both $R^2$ and **Adjusted** $R^2$ are employed as model evaluation metrics. This metric's values lie within the range of 0 to 1, where 0 signifies that the model does not explain any variability, and 1 signifies that the model perfectly explains the variability.

## Formulas:

#### Coefficient of Determination $(R^2)$:

$$\large R^2 = 1 - \frac{SS_{\text{res}}}{SS_{\text{tot}}}$$

Where:
- $SS_{\text{res}}$ represents the sum of squared residuals, which signifies the discrepancy between actual and predicted values of the dependent variable.
- $SS_{\text{tot}}$ denotes the total sum of squares, reflecting the difference between actual values of the dependent variable and the mean of the dependent variable.

#### Sum of Squared Residuals ($SS_{\text{res}}$):

$$\large SS_{\text{res}} = \sum_{i=1}^{n} (y_i - \hat{y}_i)^2$$

Where:
- $n$ is the number of data points.
- $y_i$ is the actual value of the dependent variable for the $i$-th data point.
- $\hat{y}_i$ is the predicted value of the dependent variable for the $i$-th data point.

#### Total Sum of Squares ($SS_{\text{tot}}$):

$$\large SS_{\text{tot}} = \sum_{i=1}^{n} (y_i - \bar{y})^2$$

Where:
- $n$ is the number of data points.
- $y_i$ is the actual value of the dependent variable for the $i$-th data point.
- $\bar{y}$ is the mean of the dependent variable values.

## Adjusted $R^2$:

Adjusted $R^2$ is crucial when the degrees of freedom (df) are considered.   
The formula for degrees of freedom is $n - k - 1$,  
- where $n$ stands for the number of observations, and  
- $k$ stands for the number of explanatory (X) variables, or features, in the dataset.  

As degrees of freedom decrease (due to the addition of more variables), $R^2$ only increases, not necessarily because the model improves, but due to the reduction in degrees of freedom. Adjusted $R^2$ addresses this by accounting for the effect of the number of variables, preventing the metric from misleadingly increasing as more features are added.

$$\large R_{\text{adj}}^2 = 1 - \frac{(1 - R^2) \cdot (n - 1)}{n - k - 1}$$

Where:
- $R^2$ is the coefficient of determination.
- $n$ is the number of data points.
- $k$ is the number of independent variables (features) in the model.

## Things to Note:

1. **Interpretation**: A higher $R^2$ value signifies that a larger proportion of the variance in the dependent variable is explained by the model. Yet, a high $R^2$ doesn't automatically imply a good model, as it could potentially be overfitting the data.
2. **Limitations**: In scenarios involving complex models or nonlinear relationships between variables, $R^2$ might be misleading.
3. **Adjusted $R^2$**: Adjusted $R^2$ considers the trade-off between model complexity and fit by accounting for the number of independent variables in the model. This provides a more accurate assessment of model performance.
4. **Comparative Analysis**: $R^2$ is most useful when comparing different models on the same dataset. Its utility might diminish when comparing models across diverse datasets.


## Worked example

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score

# Generating synthetic data
np.random.seed(0)
X = np.random.rand(50, 1) * 10
y = 2 * X + 1 + np.random.randn(50, 1) * 2

# Create a linear regression model
model = LinearRegression()

# Fit the model
model.fit(X, y)

# Predict the target values
y_pred = model.predict(X)

# Calculate R^2
r2 = r2_score(y, y_pred)
print("R-squared:", r2)

# Calculate the mean of the original y values
y_mean = np.mean(y)

plt.figure(figsize=(9,5))

# Plot the scatter plot of data points
plt.scatter(X, y, color='steelblue', label='Actual data')

# Plot the predicted values
plt.plot(X, y_pred, color='red', label='Regression line')

# Plot the mean of the original y values as a horizontal line
plt.axhline(y_mean, color='gray', linestyle='--', label='Mean of y')

plt.xlabel('X values')
plt.ylabel('y values')
plt.title('Linear Regression with Synthetic Data')
plt.legend()
plt.grid(True, ls='--', alpha=0.4)
plt.show()
```
